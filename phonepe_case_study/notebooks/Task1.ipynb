# Task 1: Data Loading and Understanding

## Task 1.2 - Display basic statistics and data types

```python
datasets = {
    "State_Txn and Users": state_txn_users,
    "State_TxnSplit": state_txn_split,
    "State_DeviceData": state_device,
    "District_Txn and Users": district_txn_users,
    "District Demographics": district_demo
}
```

```python
for name, df in datasets.items():

    print("\n" + "="*60)
    print(name)
    print("="*60)

    print(df.describe())   # Summary Statistics
```

```python
for name, df in datasets.items():

    print("\n" + "="*60)
    print(name)
    print("="*60)

    print(df.dtypes)    # Datatype of every column
```

## Task 1.3 - Check Missing Values

```python
for name, df in datasets.items():

    print("\n" + "="*60)
    print(name)
    print("="*60)

    print(df.isnull().sum())   # Shows number of missing values in every column.
```

```python
for name, df in datasets.items():

    print("\n" + "="*60)
    print(name)
    print("="*60)

    missing_percentage = (df.isnull().sum() / len(df)) * 100

    print(missing_percentage)   # Missing Value Percentage
```

```python
for name, df in datasets.items():

    missing_percentage = (df.isnull().sum() / len(df)) * 100

    print("\nDataset:", name)

    print("Column with highest missing percentage:")

    print(missing_percentage.idxmax())    # Highest Missing Value Column

    print("Percentage:")

    print(round(missing_percentage.max(), 2), "%")
```

## Task 1.4 - Create Summary

```python
total_states = district_demo["State"].nunique()

total_districts = district_demo["District"].nunique()

#Total States and Districts

print("Total States:", total_states)

print("Total Districts:", total_districts)
```

```python
district_count = district_demo["State"].value_counts()

state_with_max_districts = district_count.idxmax()

max_districts = district_count.max()

# State with Highest Number of Districts

print("State with Highest Number of Districts:")
print(state_with_max_districts)

print("\nNumber of Districts:")
print(max_districts)
```
